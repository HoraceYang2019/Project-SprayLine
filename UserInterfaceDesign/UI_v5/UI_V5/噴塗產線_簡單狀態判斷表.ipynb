# 噴塗產線簡單狀態判斷表

這份只整理「看哪個地方、數值多少、判斷正常 / 注意 / 異常」。


## 一、三站都用同一套判斷

底漆站、面漆站、金漆站都是看下面這些數值。

| 位置 / 零件 | 看哪個數值 | 正常 | 注意 | 異常 |
|---|---|---|---|---|
| 機械手臂 RobotArm | 可用度 | ≥ 80% | 60%～79.9% | < 60% |
| 噴嘴 Nozzle | 堵塞率 | < 30% | 30%～59.9% | ≥ 60% |
| 空壓機 AirCompressor | 壓力 | ≥ 2.0 bar | 1.5～1.9 bar | < 1.5 bar |
| 噴幅 SprayWidth | 噴幅寬度 | 在目標範圍內 | 稍微超出目標範圍 | 明顯超出目標範圍 |
| 濾網 FilterMesh | 維護性 | ≥ 50% | 25%～49.9% | < 25% |
| 品質 Quality | 膜厚穩定度 | ≥ 90% | 80%～89.9% | < 80% |


## 二、噴幅 SprayWidth 比較特別

噴幅不是固定越大越好，而是要看每站的目標範圍。

例如目標範圍是：

```text
48～56 mm
```

那就是：

```text
48～56 mm：正常
低於 48 或高於 56：不正常
偏一點：注意
偏太多：異常
```


## 三、整站狀態判斷

```text
全部零件正常 → 該站正常 / 運行中

有零件注意，但沒有異常 → 該站注意 / 維護

只要有一個零件異常 → 該站異常
```

簡單講：

```text
先看每個零件數值
再判斷正常、注意、異常
最後決定整站狀態
```
